```python
import pandas as pd
```

```python
def read_all_files():
    data_list = []
    for year in range(2003, 2016):
        df_temp = pd.read_csv(f"~/720/pds2021-opioids-team-8-1/00_source_data/US_VitalStatistics/Underlying Cause of Death, {year}.txt", delimiter="\t")
        data_list.append(df_temp)
    return pd.concat(data_list)
```

```python
df = read_all_files()
```

```python
df.sample(20)
```

```python
## Remove abnormal values from data scripting
df = df.drop(df[df["County"].isnull()].index)
```

```python
## Data Validity Check
assert not df["County"].isnull().any()
assert not df["Year"].isnull().any()
assert not df["Drug/Alcohol Induced Cause"].isnull().any()
assert not df["Deaths"].isnull().any()
```

```python
# Since I only need data about Washington, subset data
# There are 39 counties in Texas
df_WA = df[df.County.str.contains("WA")].copy()
df_WA.head(20)
```

```python
df_WA.to_csv("~/720/pds2021-opioids-team-8-1/20_intermediate_files/Underlying Cause of Death, 2003-2015, Washington.csv")
```

```python
import pandas as pd
```

```python
p_2000 = pd.read_excel("../00_source_data/texas_population_2000_2009.xls")
p_2010 = pd.read_excel("../00_source_data/texas_population_2010_2019.xlsx")
```

```python
p_2000.head()
```

```python
p_2010.head()
```

```python
## Data validity check
def validity_check(df):
    for column in df.columns:
        assert not df[column].isnull().any()
```

```python
validity_check(p_2000)
validity_check(p_2010)
```

```python
## Merger two dataset and get one intermediate file
## I decide to use Couty as the primary key, but a little bit modification need to be applied
p_2010["County"] = p_2010["County"].str.split(", ").str[0]
p_2010.head()
```

```python
TX_population = p_2000.merge(p_2010, validate="1:1", indicator=True)
assert TX_population[TX_population["_merge"] != "both"].empty
TX_population["County"] = TX_population["County"].str[1:]
TX_population.columns = TX_population.columns.astype(str)
TX_population.head()
```

```python
# TX_population.to_parquet("../20_intermediate_files/Population, 2000-2019, Texas.gzip", compression="gzip")
TX_population.to_csv("../20_intermediate_files/Population, 2000-2019, Texas.csv")
```
